```python
from ftplib import FTP
import pandas as pd

def download_csv(ftp_server, username, password, port, file_path, local_filename):
    with FTP() as ftp:
        ftp.connect(host=ftp_server, port=port)
        ftp.login(username, password)
        with open(local_filename, 'wb') as local_file:
            ftp.retrbinary(f'RETR {file_path}', local_file.write)

def read_csv(local_filename):
    # Treat empty strings as NaN
    df = pd.read_csv(local_filename, na_values=[''])
    return df

def remove_empty_sku_rows(df):
    # Removing rows where 'Variant SKU' is empty string or NaN
    return df[df['Variant SKU'].notna() & (df['Variant SKU'] != '')]



# FTP Server Details
ftp_server = 'ftp.viapu.com'
username = 'mgotest@viapu.com'
password = '35h1bn1<9Dn5'
port = 21
file_path = '/viapu.com/public_html/mgotest/cat.csv'  # Changed to CSV
local_filename = 'downloaded_cat.csv'                # Changed to CSV

# Download and Read CSV File
download_csv(ftp_server, username, password, port, file_path, local_filename)  # Changed to CSV
df = read_csv(local_filename)  # Changed to CSV



df_filtered = remove_empty_sku_rows(df)



# Display the DataFrame to verify changes
print(df_filtered)
```
